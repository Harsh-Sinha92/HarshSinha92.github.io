```python
import pandas as pd
import os as os
from openpyxl import load_workbook
from openpyxl import Workbook
from datetime import datetime,timedelta
from openpyxl.styles import Font, PatternFill, Alignment, Protection
```

```python
#Completing the operation for single file
current_dir=os.getcwd
new_dir=r'C:\\Users\\user\\Desktop'
os.chdir(new_dir)
current_dir=os.getcwd()
print("Current Directory after directory change: ",current_dir)
Files_inside_dir=os.listdir()
print("Files inside Directory: ",Files_inside_dir)
# Get the current month and year
current_date = datetime.now()
# Calculate the previous month
prev_month_date = current_date.replace(day=1)- timedelta(days=1) 
prev_month = prev_month_date.strftime("%B") 
prev_month_short = prev_month_date.strftime("%b") 
print("Name of the month: ",prev_month)
print("Name of the month: ",prev_month_short)
current_year = current_date.strftime("%Y")
current_year_short = current_date.strftime("%y")
# Construct the Destination and source filename
file_aramark = f"Labor hours - {prev_month} {current_year}.xlsx"
print("Constructed CMI filename:",file_aramark)
# Check if the constructed filenames are in the directory
if file_aramark in Files_inside_dir:
    print("All required files found in directory.")
else:
    print("One or more required files not found in directory.")
print("Once the files are read correctly process of coying and writing into JE file starts from here onwards.")
#Opening Source file from where value is to be taken
source_wb=load_workbook(file_aramark,data_only=True)
source_ws = source_wb['Journal Load Template']
sheet_JEUse = source_wb.create_sheet(title='JEUse')
# Copy the contents from the original sheet to the new sheet
for row in source_ws.iter_rows():
    for cell in row:
        new_cell = sheet_JEUse[cell.coordinate]
        new_cell.value = cell.value
        if cell.has_style:
            new_cell.font = Font(
                name=cell.font.name, 
                size=cell.font.size, 
                bold=cell.font.bold, 
                italic=cell.font.italic, 
                vertAlign=cell.font.vertAlign, 
                underline=cell.font.underline, 
                strike=cell.font.strike, 
                color=cell.font.color
            )
            new_cell.border = None
            if isinstance(cell.fill, PatternFill):
                new_cell.fill = PatternFill(
                    fill_type=cell.fill.fill_type,
                    start_color=cell.fill.start_color,
                    end_color=cell.fill.end_color
                )
            new_cell.number_format = cell.number_format
            new_cell.protection = Protection(
                locked=cell.protection.locked, 
                hidden=cell.protection.hidden
            )
            new_cell.alignment = Alignment(
                horizontal=cell.alignment.horizontal, 
                vertical=cell.alignment.vertical, 
                text_rotation=cell.alignment.text_rotation, 
                wrap_text=cell.alignment.wrap_text, 
                shrink_to_fit=cell.alignment.shrink_to_fit, 
                indent=cell.alignment.indent
            )
# Copy column widths
for col in source_ws.column_dimensions:
        sheet_JEUse.column_dimensions[col].width = source_ws.column_dimensions[col].width
# Copy row heights
for row in source_ws.row_dimensions:
        sheet_JEUse.row_dimensions[row].height = source_ws.row_dimensions[row].height
sheets = source_wb._sheets  # Access the list of sheets
sheets.insert(1, sheets.pop(-1))
print("The preparation of Aramark JE starts here.")
# Get the newly created sheet by name
sheetJEUse = source_wb['JEUse']
#Rounding up the values to 2 points of decimal.
for row in sheetJEUse.iter_rows(min_col=6, max_col=6, min_row=2):
        for cell in row:
            if isinstance(cell.value, (int, float)):  # Check if the cell contains a number
                cell.value = round(cell.value, 2)
print("Data have been rounded upto 2 points of decimal.")
# Filter out rows with 0 values in 'Units' column (column F) and delete them
rows_to_delete = []
for row in sheetJEUse.iter_rows(min_row=2):  # Assuming first row is the header
    units_cell = row[5]  # Column F is the 6th column (0-indexed)
    if units_cell.value == 0 or units_cell.value == '-':
        rows_to_delete.append(row[0].row)  # Mark the row number for deletion
# Delete rows in reverse order to avoid messing up the indexing
for row_num in sorted(rows_to_delete, reverse=True):
    sheetJEUse.delete_rows(row_num, 1)
print(f"Deleted {len(rows_to_delete)} rows with blank values in the 'Units' column.")
filter_values = [
    (2020,46900)
]
# Numbers to write in columns G and H for the matching rows
activity_number = 46502362000
# Apply the filter and write numbers in columns G and H
for row in sheetJEUse.iter_rows(min_row=2):
    if (row[0].value, row[1].value) in filter_values:
        row[6].value = activity_number
print("Activity number: 46502362000 have been updated for 2020-46900.")
for row in sheetJEUse.iter_rows(min_row=2):  # Assuming first row is the header
    if row[0].value == 5141 and row[1].value == 85045:
        row[0].value = 5140
print("Updated rows where CO/AU combination are 5141-85045 to 5140-85045.")
# Save the workbook after performing operations
source_wb.save(file_aramark)
print("Aramark JE prepared.")
```
