```python
import pandas as pd
import os as os
from openpyxl import load_workbook
from openpyxl import Workbook
from datetime import datetime,timedelta
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
#Completing the operation for single file
current_dir=os.getcwd
new_dir=r'C:\\Users\\user\\Desktop'
os.chdir(new_dir)
current_dir=os.getcwd()
print("Current Directory after directory change: ",current_dir)
Files_inside_dir=os.listdir()
print("Files inside Directory: ",Files_inside_dir)
# Get the current month and year
current_date = datetime.now()
# Calculate the previous month
prev_month_date = current_date.replace(day=1)- timedelta(days=1) 
prev_month = prev_month_date.strftime("%B") 
prev_month_short = prev_month_date.strftime("%b") 
prev_month_short_caps=prev_month_short.upper()
print("Name of the month in caps: ",prev_month_short_caps)
current_year = current_date.strftime("%Y")
current_year_short = current_date.strftime("%y")
# Create a new date object with the current month, a fixed day (7th), and the current year
dynamic_date = datetime(current_date.year, current_date.month, 7)
# Format the date as mmddyyyy
formatted_date = dynamic_date.strftime('%m%d%Y')
print("Date in format: ",formatted_date)
# Construct the Destination and source filename
file_CMI = f"CMI & CWAD Calculations {prev_month} {current_year}.xlsx".strip()
file_dynamic_CMI=f"950680_{prev_month_short_caps}_{formatted_date}.xlsx"
file_by_payor=f"MIDAS CMI By Payor {prev_month_short_caps}.xlsx"
file_d = f"CMICWADMOYTD.xlsx"
print("Constructed CMI filename:",file_CMI)
print("Constructed Dynamic CMI filename:",file_dynamic_CMI)
print("Constructed By Payor CMI filename:",file_by_payor)
print("Constructed Destination filename:",file_d)
#files_to_check=[file_by_payor,file_dynamic_CMI,file_CMI]
# Check if the constructed filenames are in the directory
if file_CMI in Files_inside_dir:
    print("CMI file found in directory.")
else:
    print("CMI file not found in directory.")
if file_dynamic_CMI in Files_inside_dir:
    print("Dynamic CMI file found in directory.")
else:
    print("Dynamic CMI file not found in directory.")
if file_by_payor in Files_inside_dir:
    print("By Payor CMI file found in directory.")
else:
    print("By Payor CMI file not found in directory.")
if file_d in Files_inside_dir:
    print("CMICWADMOYTD found in directory:")
else:
    print("CMICWADMOYTD not found in directory.")
print("Once the files are read correctly process of coying and writing into JE file starts from here onwards.")
# Read data from source Excel file 1
source_file_CMI=file_CMI
source_sheet_CMI="OrigJEsInfo"
source_df_CMI = pd.read_excel(source_file_CMI, sheet_name=source_sheet_CMI, na_values=[''])  # Treat empty strings as missing values
# Drop any rows with all NaN values
source_df_CMI.dropna(how='all', inplace=True)
# Extract data from the 2nd column onwards
data_to_write_CMI = source_df_CMI.iloc[:, 1:].values.tolist()
# Read data from source Excel file 2
source_file_dynamic_CMI=file_dynamic_CMI
source_sheet_dynamic_CMI="950680_Current Mnth JE"
source_df_dynamic_CMI = pd.read_excel(source_file_dynamic_CMI, sheet_name=source_sheet_dynamic_CMI, na_values=[''])
# Drop any rows with all NaN values
source_df_dynamic_CMI.dropna(how='all', inplace=True)
# Extract data from the 2nd column onwards
data_to_write_dynamic_CMI = source_df_dynamic_CMI.iloc[:, 0:].values.tolist()
# Read data from source Excel file 3
source_file_by_payor=file_by_payor
source_sheet_by_payor="cmi by payer"
source_df_by_payor = pd.read_excel(source_file_by_payor, sheet_name=source_sheet_by_payor, na_values=[''])
# Drop any rows with all NaN values
source_df_by_payor.dropna(how='all', inplace=True)
# Extract data from the 33rd column onwards
data_to_write_by_payor = source_df_by_payor.iloc[1:, 32:].values.tolist()
# Load the destination workbook
destination_file=file_d
wb = load_workbook(destination_file)
# Select the active worksheet
ws = wb.active
ws.title="Input"
# Set the start row to 13
start_row = 13
# Find the next empty row in the destination file
#start_row = ws.max_row + 1
# Write the data from source file 1 into the destination file
for row_data in data_to_write_CMI:
    for col_index, value in enumerate(row_data, start=1):
        ws.cell(row=start_row, column=col_index, value=value)
    start_row += 1
# Write the data from source file 2 into the destination file
for row_data in data_to_write_dynamic_CMI:
    for col_index, value in enumerate(row_data, start=1):
        ws.cell(row=start_row, column=col_index, value=value)
    start_row += 1
# Write the data from source file 3 into the destination file
for row_data in data_to_write_by_payor:
    for col_index, value in enumerate(row_data, start=1):
        ws.cell(row=start_row, column=col_index, value=value)
    start_row += 1
# Save the workbook back to the destination file
wb.save(destination_file)
print(f"Data successfully written to {destination_file}")
```
