```python
import pandas as pd
import os as os
from openpyxl import load_workbook
from datetime import datetime,timedelta
```

```python
print("Data received from Courtney McGill.")
# File paths
current_dir=os.getcwd
new_dir=r'C:\\Users\\user\\Desktop'
os.chdir(new_dir)
current_dir=os.getcwd()
print("Current Directory after directory change: ",current_dir)
Files_inside_dir=os.listdir()
print("Files inside Directory: ",Files_inside_dir)
# Get the current month and year
current_date = datetime.now()
# Calculate the previous month
prev_month_date = current_date.replace(day=1)- timedelta(days=1) 
prev_month = prev_month_date.strftime("%B") 
prev_month_short = prev_month_date.strftime("%b") 
current_year = current_date.strftime("%Y")
current_year_short = current_date.strftime("%y")
# Add one more year to the current date
next_year_date = current_date.replace(year=current_date.year + 1)
next_year = next_year_date.strftime("%Y")
print("Next Year: ", next_year)
# Construct the Destination and source filename
file_s = f"Laundry_MASTER_FY{next_year}.xlsx" #Courtney McGill
file_d = f"{prev_month_short} {current_year} Consolidated Stats.xlsx"
print("Constructed Source filename:", file_s)
print("Constructed Destination filename:", file_d)
# Check if the constructed filenames are in the directory
if file_s in Files_inside_dir:
    print("Source filename found in directory:", file_s)
else:
    print("Source filename not found in directory.")
if file_d in Files_inside_dir:
    print("Destination filename found in directory:", file_d)
else:
    print("Destination filename not found in directory.")
# Open source and destination files
source_wb = load_workbook(file_s, data_only=True)
destination_wb = load_workbook(file_d)
# Calculate previous month and year
previous_month_date = current_date - timedelta(days=current_date.day)
desired_month = previous_month_date.strftime("%b")
desired_year_suffix = previous_month_date.strftime("%y")
desired_month_year = f"{desired_month}-{desired_year_suffix}"
print("Current Date: ", current_date)
print("Desired Month-Year: ", desired_month_year)
# Mapping months to column letters (C5 to N5 for Jul-23 to Jun-24)
month_column_mapping = {
    'Jul-24': 'C', 'Aug-24': 'D', 'Sep-24': 'E', 'Oct-24': 'F',
    'Nov-24': 'G', 'Dec-24': 'H', 'Jan-25': 'I', 'Feb-25': 'J',
    'Mar-25': 'K', 'Apr-25': 'L', 'May-25': 'M', 'Jun-25': 'N'
}
print("Month Column Mapping: ", month_column_mapping)
# Determine the column letter based on the previous month
value_column_letter = month_column_mapping.get(desired_month_year)
print("Value Column Letter: ", value_column_letter)
if value_column_letter:
    # Sheet and IDs mapping
    sheets_info = {
        'Plano Heart Hospital': {'ids': (1131, 81710, 900500), 'source_row': 39},
        'Denton Heart Hospital': {'ids': (1133, 81710, 900500), 'source_row': 39},
        'McKinney Remote Heart Hospital': {'ids': (1134, 81710, 900500), 'source_row': 39}
    }
    for sheet_name, info in sheets_info.items():
        source_ws = source_wb[sheet_name]
        destination_ws = destination_wb['Journal Load Template']
        id_1, id_2, id_3 = info['ids']
        source_row = info['source_row']
        des_row = None
        # Find the row to write the value
        for row in destination_ws.iter_rows(min_row=2, max_row=destination_ws.max_row, min_col=1, max_col=3):
            if row[0].value == id_1 and row[1].value == id_2 and row[2].value == id_3:
                des_row = row[0].row
                break
        # Check if the destination row is found
        if des_row:
            print(f"Destination Row Found in {sheet_name}: ", des_row)
            # Check if the date in the determined column matches the previous month and year
            for cell in source_ws[5]: 
                if cell.value and cell.value.upper() == desired_month_year.upper():
                    source_value_column = cell.column
                    source_value = source_ws.cell(row=source_row, column=source_value_column).value
                    print("Column:", source_value_column)
                    print("Source Value:", source_value)
                    # Write the value to the destination cell
                    destination_ws.cell(row=des_row, column=6).value = source_value 
                    print("Value written to destination cell in", sheet_name)
                    break
            else:
                print(f"Desired month not found in source file {sheet_name}.")
        else:
            print(f"Destination row not found in {sheet_name}.")
else:
    print("Value Column Letter not found.")
# Save and close files
destination_wb.save(file_d)
destination_wb.close()
source_wb.close()
print("Files have been updated successfully for accounts 1131,1133 and 1134 respectively.")
```
